```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("global_energy_mix.csv")

fossil_sources = ['Coal', 'Oil', 'Natural Gas']

fossil_df = df[df['Source'].isin(fossil_sources)]
```

```python
fig = px.treemap(
    fossil_df,
    path=['Region', 'Country', 'Source'],
    values='TWh',
    color='Source',
    color_discrete_sequence=['red', 'orange', 'brown'],
    title='Fossil Fuel Dependency by Country'
)

fig.update_traces(
    textinfo='label+value+percent parent'
)

fig.show()
```

**TASK 2**

```python
import plotly.express as px
```

```python
tips = px.data.tips()

tips_summary = (
    tips.groupby(['day', 'time', 'smoker'])['total_bill']
    .sum()
    .reset_index()
)
```

```python
fig = px.sunburst(
    tips_summary,
    path=['day', 'time', 'smoker'],
    values='total_bill',
    color='time',
    color_discrete_sequence=['blue', 'green'],
    title='Total Bill Distribution by Day, Time and Smoker Status'
)

fig.update_traces(
    textinfo='label+percent parent'
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("global_energy_mix.csv")

low_carbon_sources = ['Nuclear', 'Hydro', 'Wind', 'Solar']

low_carbon_df = df[df['Source'].isin(low_carbon_sources)]
```

```python
country_energy = (
    low_carbon_df.groupby('Country')['TWh']
    .sum()
    .reset_index()
    .sort_values(by='TWh', ascending=False)
)
```

```python
fig1 = px.treemap(
    country_energy,
    path=['Country'],
    values='TWh',
    color='TWh',
    title='Low-Carbon Energy Production by Country (Treemap)'
)
fig1.show()
```

```python
fig2 = px.bar(
    country_energy,
    x='TWh',
    y='Country',
    orientation='h',
    color='TWh',
    color_continuous_scale='viridis',
    title='Low-Carbon Energy Production by Country (Bar Chart)'
)

fig2.show()
```

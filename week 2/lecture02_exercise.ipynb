```python
import pandas as pd

df = pd.read_csv("world_happiness_2023.csv")
df.head()
```

```python
df.columns = df.columns.str.strip()
region_avg = (df.groupby('Regional indicator')['Ladder score']
              .mean()
              .reset_index()
              .sort_values('Ladder score', ascending=True))
import plotly.express as px

fig = px.bar(
    region_avg,
    x='Ladder score',
    y='Regional indicator',
    orientation='h',
    text='Ladder score',
    color='Ladder score',
    color_continuous_scale='Viridis',
    title='Which region stands out in happiness, and why does it matter?'
)

fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')

fig.update_layout(
    xaxis_title='Average Happiness (Ladder Score)',
    yaxis_title='Region',
    xaxis=dict(range=[0, region_avg['Ladder score'].max() + 1]),
    yaxis={'categoryorder': 'total ascending'}
)

fig.show()
```

```python
import pandas as pd

# Top 8 happiest countries
top8 = df.nlargest(8, 'Ladder score').copy()
top8['Group'] = 'Top 8'

# Bottom 8 least happy countries
bottom8 = df.nsmallest(8, 'Ladder score').copy()
bottom8['Group'] = 'Bottom 8'

# Combine both
combined = pd.concat([bottom8, top8])

# Global average
global_avg = df['Ladder score'].mean()
print("Global average:", round(global_avg, 2))
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.bar(
    combined,
    x='Ladder score',
    y='Country name',
    color='Group',
    orientation='h',
    text='Ladder score',
    color_discrete_map={
        'Top 8': '#3776ab',      # blue
        'Bottom 8': '#37ab3c'    # green
    },
    title='Extreme Happiness Gap: Top vs Bottom Countries'
)

# Add value labels
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')

# Add global average line
fig.add_vline(
    x=global_avg,
    line_width=2,
    line_dash="dash",
    line_color="blue"
)

# Layout improvements
fig.update_layout(
    xaxis_title="Happiness Score (Ladder Score)",
    yaxis_title="Country",
    xaxis=dict(range=[0, df['Ladder score'].max() + 1]),
    legend_title="Group"
)
```

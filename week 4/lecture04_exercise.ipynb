```python
df_2007 = df[df['year'] == 2007].copy()

# Assign highlight group: Asia in color, everything else grey
df_2007['highlight'] = df_2007['continent'].apply(
    lambda c: 'Asia' if c == 'Asia' else 'Other'
)
```

```python
df_2007 = df_2007.sort_values('highlight', ascending=False)

fig2 = px.scatter(
    df_2007,
    x='gdpPercap',
    y='lifeExp',
    size='pop',
    color='highlight',
    color_discrete_map={'Asia': '#e63946', 'Other': '#cccccc'},
    hover_name='country',
    hover_data={'pop': ':,', 'gdpPercap': ':,.0f', 'highlight': False},
    size_max=55,
    log_x=True,
    labels={
        'gdpPercap': 'GDP per Capita (log scale, USD)',
        'lifeExp':   'Life Expectancy (years)',
        'pop':       'Population',
        'highlight': ''
    },
    title="Asia in 2007: a continent of extremes — from poverty to prosperity",
)
```

```python
fig2.add_annotation(
    x=4.15, y=73,
    xref='x', yref='y',
    text='Asia spans the full spectrum —<br>rich city-states to low-income nations',
    showarrow=True,
    arrowhead=2,
    arrowcolor='#e63946',
    font=dict(size=11, color='#e63946'),
    ax=-140, ay=-40,
    bgcolor='white',
    bordercolor='#e63946',
    borderwidth=1,
)
```

```python
fig2.update_layout(
    plot_bgcolor='white',
    paper_bgcolor='white',
    font_family='Arial',
    title_font_size=15,
    legend=dict(itemsizing='constant', title_text='')
)
fig2.update_xaxes(showgrid=True, gridcolor='#eeeeee')
fig2.update_yaxes(showgrid=True, gridcolor='#eeeeee')

fig2.show()
```

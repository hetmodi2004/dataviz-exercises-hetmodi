## **Task 2 **

```python
df_slope = df[df['Year'].isin([2000, 2022])]
df_avg = df_slope.groupby(['Region', 'Year'])['CO2_Mt'].mean().reset_index()
pivot = df_avg.pivot(index='Region', columns='Year', values='CO2_Mt').dropna()
pivot
```

```python
import plotly.graph_objects as go

fig2 = go.Figure()

for region in pivot.index:
    fig2.add_trace(go.Scatter(
        x=[2000, 2022],
        y=[pivot.loc[region, 2000], pivot.loc[region, 2022]],
        mode='lines+markers',
        name=region,
        line=dict(width=2),
        marker=dict(size=6),
        hovertemplate=f"{region}<br>Year: %{{x}}<br>Value: %{{y}}<extra></extra>"
    ))

fig2.update_layout(
    title="Change from 2000 to 2022 by Region",
    xaxis_title="Year",
    yaxis_title="Value",
    template="plotly_white",
    legend_title="Region",
    hovermode="x unified"
)

fig2.show()
```
